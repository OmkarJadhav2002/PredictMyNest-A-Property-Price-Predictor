```python
import pandas as pd 
import numpy as np 
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
house_df = pd.read_csv("Datasets/houses.csv")
house_df.head()
```

```python
# check the number of rows and number of columns 
house_df.shape
```

```python
# check for null values 
house_df.isna().sum()
```

```python
# check for duplicate values 
house_df[house_df.duplicated()]
```

```python
# drop the duplicates 
house_df = house_df.drop_duplicates()
```

```python
# check for missing values
house_df.isnull().sum()
```

```python
# Columns to drop -> property_name, link, property_id
house_df.drop(columns=['link','property_id'], inplace=True)
```

```python
house_df.head()
```

```python
# rename columns
house_df.rename(columns={'rate':'price_per_sqft'},inplace=True)
```

```python
# cleaning the society column 
house_df["society"].value_counts()
```

```python
house_df["society"].value_counts().shape
```

```python
import re 
house_df['society'] = house_df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
house_df["society"].value_counts().shape
```

```python
house_df['society'] = house_df['society'].str.replace('nan','independent')
```

```python
# cleaning the price column and scale down to same scale (crore)
house_df["price"].value_counts()
```

```python
house_df = house_df[house_df["price"] != "Price on Request"]
```

```python
def scale_price(p):
    if type(p) == float:
        return p 
    else:
        if p[1] == "Lac":
            return round(float(p[0])/100, 2) 
        else:
            return round(float(p[0]), 2)
```

```python
house_df["price"] = house_df["price"].str.split(' ').apply(scale_price)
```

```python
house_df.head()
```

```python
# cleaning the price_per_sqft column 
house_df["price_per_sqft"] = house_df["price_per_sqft"].str.split(' ').str.get(1).str.split('/').str.get(0).str.replace(',', '').str.strip().astype("float")
```

```python
# cleaning the bedRoom column 
house_df["bedRoom"].value_counts()
```

```python
# check for null values 
house_df[house_df["bedRoom"].isnull()]
```

```python
# drop the rows where bedRoom is NaN 
house_df = house_df[~house_df["bedRoom"].isnull()]
```

```python
house_df.shape
```

```python
# now clean the column 
house_df["bedRoom"] = house_df["bedRoom"].str.split().str.get(0).astype("int")
```

```python
# cleaning the bathroom column 
house_df["bathroom"].isnull().sum() # there is no missing value 
```

```python
house_df["bathroom"].value_counts()
```

```python
house_df["bathroom"] = house_df["bathroom"].str.split().str.get(0).astype("int")
```

```python
# cleaning the balcony column 
house_df["balcony"].value_counts()
```

```python
# check for null values 
house_df["balcony"].isnull().sum() # there are no null values 
```

```python
house_df["balcony"] = house_df["balcony"].str.split().str.get(0).str.replace('+', '').str.replace('No', '0').astype("int")
```

```python
house_df["balcony"].value_counts()
```

```python
# cleaning additionalRoom column (this column needs feature engineering)
house_df["additionalRoom"].value_counts()
```

```python
# check for null values 
house_df["additionalRoom"][house_df["additionalRoom"].isnull()]
```

```python
# fill the NaN with not available 
house_df["additionalRoom"].fillna("not available", inplace=True)
```

```python
house_df["additionalRoom"][house_df["additionalRoom"] == "not available"]
```

```python
# convert all the information to lower case 
house_df['additionalRoom'] = house_df['additionalRoom'].str.lower()
```

```python
# cleaning the noOffloor column 
house_df["noOfFloor"].isnull().sum()
```

```python
house_df["noOfFloor"] = house_df["noOfFloor"].str.split().str.get(0)
```

```python
# rename the column noOffloor 
house_df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
# cleaning the facing column 
house_df['facing'].fillna('NA',inplace=True)
```

```python
# adding the extra column area from price and price_per_sqft
house_df['area'] = round((house_df['price']*10000000)/house_df['price_per_sqft'])
```

```python
# insert the new column which tells the property type 
house_df.insert(loc=1,column='property_type', value='house')
```

```python
# export the dataset 
house_df.to_csv("house_cleaned_phase1.csv", index=False)
```
