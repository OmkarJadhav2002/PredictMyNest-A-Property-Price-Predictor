```python
import pandas as pd 
```

```python
flats = pd.read_csv("flats_cleaned_phase1.csv") 
houses = pd.read_csv("house_cleaned_phase1.csv") 
```

```python
flats.head()
```

```python
houses.head()
```

```python
# concat the above two datasets 
merged_df = pd.concat([flats, houses], ignore_index=True)
```

```python
# shuffle the data 
shuffled_df = merged_df.sample(merged_df.shape[0], ignore_index=True)
```

```python
shuffled_df.sample(3)
```

```python
# export the dataset (this is the final dataset after phase1 cleaning) 
shuffled_df.to_csv("gurgaon_properties.csv", index=False)
```
